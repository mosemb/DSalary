```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('CleanedDf.csv')
```

```python
df.head()
```

```python
df['CompanyName'] = df['CompanyName'].replace({'Buchanan And Edwards Group\n4.0':'Buchanan And Edwards Group'})
```

```python
df.CompanyName.value_counts()
```

```python
df.drop(columns=df.columns[0], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.drop(['Aws', 'Sql', 'MachineLearning', 'sql', 'Spark'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.drop(['HourlymaxSal'], axis=1, inplace=True)	
```

```python
df.head()
```

```python
df.RevenueType.value_counts()
```

```python
df['RevenueType'] = df['RevenueType'].replace({'-2':'Unknown'})
```

```python
df.RevenueType.value_counts()
```

```python
df.drop(['minSalary', 'maxSalary'], axis=1, inplace=True)	
```

```python
df.head()
```

```python
df.Job_Title.value_counts()
```

```python
df['Job_Title'] = df['Job_Title'].replace({'Data Scientist 4676':'Data Scientist'})
```

```python
df.Job_Title.value_counts()
```

```python
df['Job_Title'] = df['Job_Title'].replace({'Entry Level Data Science - December 2023 and May 2024 Grads':'Data Scientist'})
```

```python
df.Job_Title.value_counts()
```

```python
df["Job_Desc_len"] = df["Job_Desc"].apply( lambda x: len(x) )
```

```python
df.head()
```

```python
#df.Job_Desc_len.head(30)
```

```python
df.describe()
```

```python
df.maxSalary_new.value_counts()
```

```python
df[df['maxSalary_new'] == 0 ]
```

```python
df['maxSalary_new'] = np.where(df['maxSalary_new'] == 0, df['minSalary_new'], df['maxSalary_new'])
```

```python
df[df['maxSalary_new'] == 0 ]
```

```python
df.describe()
```

```python
#Average Salary
df["AverageSalary"] = df[["minSalary_new","maxSalary_new"]].mean(axis=1)
```
